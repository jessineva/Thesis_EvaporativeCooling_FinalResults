```python
import iertools.read as rd
import pandas as pd
```

```python
f = '../data/Casablanca.epw'
epw =  rd.read_epw(f)
epw.plot(subplots=True, figsize=(20,20));
```

```python
f = '../data/Casablanca.epw'
epw =  rd.read_epw(f,alias=True,year=1960)
epw.plot(subplots=True, figsize=(20,20));
```

```python
epw_To = epw.copy()
epw_To.To = 20

epw.To.plot(figsize=(20,3))
epw_To.To.plot(figsize=(20,3))
```

```python
rd.to_epw('../data/CasablancaTo20.epw',epw_To,'../data/Casablanca.epw')
```

```python
epw_To_written = rd.read_epw('../data/CasablancaTo20.epw',alias=True)
epw_To_written.plot(subplots=True, figsize=(20,20));
```
