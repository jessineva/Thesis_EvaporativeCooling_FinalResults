```python
import pandas as pd
import psychrolib as ps #DOI  10.5281/zendi.2537945
ps.SetUnitSystem(ps.SI)
import numpy as np
import glob as glob
import matplotlib.pyplot as plt
```

```python
def epw(inicio, final, freq, To, HR, ws, wd, masl=1247):
    
    data      = pd.DataFrame(index=pd.date_range(inicio, final, freq=freq), columns=["year", "month", "day",
                                                                                     "hour", "minute",
                                                                                     "To", "HR", "Patm","ws", "wd"])
    data.year   = data.index.year
    data.month  = data.index.month
    data.day    = data.index.day
    data.hour   = data.index.hour
    data.minute = data.index.minute
    data.To     = To
    data.HR     = HR
    data.Patm   = ps.GetStandardAtmPressure(masl)
    data.ws     = ws
    
    deg       = [0, 90, 180, 270]
    direction = ["North", "East", "South", "West"]
    
    for i in np.arange(4):
        if direction[i] == wd:
            data.wd  = deg[i]
            
    data.drop(data.index[-1], inplace=True)
    return data
```

## Saltos en epw

1. Todo igual
2. velocidades de viento diferente 
3. humedad relativa diferente
4. temperaturas diferentes
5. temperaturas diferentes y humedad relativas diferente


## Todo cero

```python
inicio = "2018-01-01 00:00:00"
final  = "2018-01-20 00:00:00"
freq   = "1min"
To     = 0
HR     = 0
ws     = 0
wd     = "West"
data1  = epw(inicio, final, freq, To, HR, ws, wd)

inicio = "2018-01-20 00:00:00"
final  = "2018-02-01 00:00:00"
freq   = "1min"
To     = 0
HR     = 0
ws     = 0
wd     = "West"
data2  = epw(inicio, final, freq, To, HR, ws, wd)

data = pd.concat([data1, data2],axis=0)
data.to_csv(f"../data/epw_TodoCero.csv",header=False, index=False)
data
```
